# Data Extraction

```python
# Dependencies
import requests
from config import api_key
from pprint import pprint
import pandas as pd
import numpy as np
```

```python
# Set up veriable for API call, year 2010 ~ 2011
Year_list=[2010,2011]
movies_df=movies_df[movies_df['year'].isin(Year_list)]
movie_list=movies_df['name'].to_list()
```

```python
# Run the API call from 'www.omdbapi.com'
url = "http://www.omdbapi.com/?apikey=" + api_key + "&t="

responses = []

try:
    for movie in movie_list:
        movie_data = requests.get(url + movie).json()
        responses.append(movie_data)
except:
    print('not found')
```

```python
pprint(responses[0])
```

```python
# Check the recieved dataframe
movies_df = pd.DataFrame(responses)
movies_df.head()
```

```python
# Save as csv file to resources
movies_df.to_csv("resources/movies_2010_2011.csv")
```
